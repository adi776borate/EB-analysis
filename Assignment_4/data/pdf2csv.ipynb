```python
import fitz
import pandas as pd
import mysql.connector 
from sqlalchemy import create_engine
```

```python
def pdf2df(inpdf):
    doc = fitz.open(inpdf)

    pdf=[]
    for i in range(doc.page_count):
        page = doc[i]
        tabs = page.find_tables()
        tab = tabs[0].to_pandas()
        
        pdf.append(tab)
        
        print(f'Page {i} processed out of {doc.page_count} pages')

        df = pd.concat(pdf)
        df=df
    return df
```

```python
df1 = pdf2df("EB_Redemption_Details.pdf")
df1.to_csv('EB_Redemption_Details.csv',index=False)
```

```python
df2 = pdf2df("EB_Purchase_Details.pdf")
df2.to_csv('EB_Purchase_Details.csv',index=False)
```
